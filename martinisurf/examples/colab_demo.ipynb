## 👀 5. Optional: Visualize in Colab (PyVista)

```python
import py3Dmol

gro_file = open("/content/MartiniSurf/Simulation_Files/2_system/immobilized_system.gro").read()

view = py3Dmol.view(width="100%", height=1000)   # 👈 ancho completo
view.addModel(gro_file, "gro")


view.setStyle({"sphere":{"radius": 1.5}})


# Anchor
view.setStyle(
    {"resi": [ 8, 10, 11, 1025, 1027, 1028]},
    {"sphere": {"color": "red", "radius": 2.5}}
)

view.setProjection("orthographic")
view.zoomTo()
view.show()
```
