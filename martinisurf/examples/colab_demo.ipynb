## 🧱 3. Build a Martini Surface
This example generates a 20×20 nm surface with P4 symmetry.

```python
!martinisurf surface \
    --lx 20 --ly 20 \
    --symmetry P4 \
    --outfile surface.gro
```

## 🧭 4. Orient an Enzyme Using Anchor Residues
Upload an enzyme PDB first (Colab panel → Upload file), then run:

```python
!martinisurf orient \
    enzyme.pdb \
    --surface surface.gro \
    --anchor 10 12 14 \
    --anchor 1025 1026 1028 \
    --outfile oriented_enzyme.pdb
```

## 🏗️ 5. Build a Full Martini + GōMartini System

```python
!martinisurf system \
    --enzyme oriented_enzyme.pdb \
    --surface surface.gro \
    --gomartini \
    --output Sim_Example
```

## 👀 6. Optional: Visualize in Colab (PyVista)

```python
import pyvista as pv
mesh = pv.read('surface.gro')
plotter = pv.Plotter(notebook=True)
plotter.add_points(mesh.points)
plotter.show()
```
